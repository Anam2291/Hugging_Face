```python
# import pipeline from transformers
from transformers import pipeline
```

# **Text classification**

```python
# create a classifier pipeline
classifier = pipeline("sentiment-analysis")
```

```python
# classify a text
text = 'I am very happy today'
result = classifier(text)
print(result)
```

```python
text1 = 'I am very happy today , but i am a bit sad too!'
result1 = classifier(text1)
print(result1)
```

```python
text2 = 'I was sad but now i am happy'
result2 = classifier(text2)
print(result2)
```

```python
text3 = "World's most advanced satellite internet.Starlink delivers high-speed internet capable of streaming, gaming, video calls and more"
result3 = classifier(text3)
print(result3)
```

# **Text generation**

```python
#  Take a generator
generator = pipeline(task="text-generation")
```

```python
# Generate a text using generator
text = generator("The world is a")
print(text[0]['generated_text'])
```

```python
# Generate a text using generator
text = generator("The world is a", max_length=50, do_sample=False, truncation=True)
print(text[0]['generated_text'])
```
